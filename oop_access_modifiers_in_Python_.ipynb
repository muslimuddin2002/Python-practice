```python
# Public Access Modifier
class Student:

    def __init__(self):
        self.name = "Muslim"   # Public variable

    def show(self):
        print("Student Name:", self.name)


s1 = Student()

# Access from outside the class
print(s1.name)

s1.show()
```

```python
# Protected Access Modifier
class Student:

    def __init__(self):
        self._department = "CSE"   # Protected variable

    def show(self):
        print("Department:", self._department)


s1 = Student()

# Accessing protected variable
print(s1._department)

s1.show()
```

```python
# Protected with Child Class Example
class Parent:

    def __init__(self):
        self._money = 50000000


class Child(Parent):

    def display(self):
        print("Father's Money:", self._money)


c1 = Child()

c1.display()
```

```python
# Private Access Modifier
class Student:

    def __init__(self):
        self.__password = "12345"   # Private variable

    def show_password(self):
        print("Password:", self.__password)


s1 = Student()

# Access through method
s1.show_password()
```

```python
# Private Variable Direct Access করলে Error
class Student:

    def __init__(self):
        self.__password = "12345"


s1 = Student()

print(s1.__password)
```

```python
# access modifiers in Python.
class Person:

    def __init__(self):

        # Public
        self.name = "Muslim"

        # Protected
        self._age = 23

        # Private
        self.__password = "muslim1234"

    def show_private(self):
        print("Private Password:", self.__password)


p1 = Person()

# Public
print(p1.name)

# Protected
print(p1._age)

# Private through method
p1.show_private()
```
