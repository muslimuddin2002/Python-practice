```python
#Student Result Management System
def calculate_gpa(avg):
    if avg >= 80:
        return 5.0
    elif avg >= 70:
        return 4.0
    elif avg >= 60:
        return 3.5
    elif avg >= 50:
        return 3.0
    elif avg >= 40:
        return 2.0
    else:
        return 0.0


def add_student():
    name = input("Enter student name: ")
    roll = input("Enter roll number: ")

    subjects = ["Bangla", "English", "Math","BDS"]
    marks = []

    for sub in subjects:
        m = float(input(f"Enter marks for {sub}: "))
        marks.append(m)

    total = sum(marks)
    average = total / len(subjects)
    gpa = calculate_gpa(average)
    result = "Pass" if gpa > 0 else "Fail"

    with open("result.txt", "a") as file:
        file.write(f"{name},{roll},{total},{average:.2f},{gpa},{result}\n")

    print("\n‚úÖ Result Saved Successfully!")
    print("Name:", name)
    print("Roll:", roll)
    print("GPA:", gpa)
    print("Result:", result)


def view_results():
    print("\nüìÑ All Student Results:\n")
    try:
        with open("result.txt", "r") as file:
            for line in file:
                name, roll, total, avg, gpa, result = line.strip().split(",")
                print(f"Name: {name} | Roll: {roll} | GPA: {gpa} | Result: {result}")
    except FileNotFoundError:
        print("No result found!")


while True:
    print("\nüéì Student Result Management System")
    print("1. Add Student Result")
    print("2. View All Results")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_results()
    elif choice == "3":
        print("üëã Exiting...")
        break
    else:
        print("‚ùå Invalid choice!")
```
