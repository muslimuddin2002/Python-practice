```python
# Stack in Python
# LIFO → Last In First Out
# Push Operation
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print(stack)
```

```python
# Pop Operation
stack = [10, 20, 30]

removed = stack.pop()

print("Removed:", removed)
print("Stack:", stack)
```

```python
# Peek / Top Element
stack = [10, 20, 30]

print(stack[-1])
```

```python
# Stack Real Life Project 🔥
# Browser History
history = []

history.append("Google")
history.append("YouTube")
history.append("Facebook")

print("Current:", history)

history.pop()

print("After Back:", history)
```

```python
# Queue in Python
# FIFO → First In First Out
# Enqueue Operation
from collections import deque

queue = deque()

queue.append(10)
queue.append(20)
queue.append(30)

print(queue)
```

```python
# Dequeue Operation
from collections import deque

queue = deque([10, 20, 30])

removed = queue.popleft()

print("Removed:", removed)
print("Queue:", queue)
```

```python
# Queue Real Life Example 🔥
# Ticket Counter System
from collections import deque

customers = deque()

customers.append("Rahim")
customers.append("Karim")
customers.append("Sakib")

print(customers)

customers.popleft()

print("After Service:", customers)
```
