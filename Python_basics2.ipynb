```python
#python Nonlocal variables
# outside function
def outer():
    message = 'local'
    #nested function
    def inner():
      #declare nonlocal variable
      nonlocal message

      message = 'nonlocal'
      print('Inner: ',message)
    inner()
    print('outer: ',message)
outer()
```

```python
from os import truncate
## Program to Check if a Number is prime
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, int(n**0.5)+1):
    if n % i == 0:
        return False
  return True
# Example usage
number = 20
print(f'Is {number} prime? {is_prime(number)}')

```

```python
from binascii import b2a_base64
# Program to Calculate the Greatest Common Divisor(GCD)

def gcd(a,b):
    while b:
      a,b=b,a % b
    return a

#Example usage
num1,num2 = 48 ,18
print(f'GCD of {num1} and {num2} is {gcd(num1, num2)}')

```

```python
# Program to Find the sum Digits of a Number

def sum_of_digits(n):
  total = 0
  while n > 0:
    total += n % 10
    n //= 10
  return total
#Example usage
number = 1234
print(f'Sum of digits of {number} is {sum_of_digits(number)}')
```

```python
# Program to Find the Longest Word in a Sentence
def longest_word(Sentence):
    words = Sentence.split()
    longest = max(words,key=len)
    return longest

#Example usage
Sentence = 'The quick brown fox jumps over the lazy dog'
print(f"The longest word in the sentence is:'{longest_word(Sentence)}'")
```

```python
## String Functions

# Initial string
s = "Skill Morph"

# 1. upper()
print(s.upper())  # Converts all characters to uppercase
# 2. lower()
print(s.lower())  # Converts all characters to lowercase
# 3. partition()
print(s.partition(' '))  # Splits the string at the first occurrence of ' '
# 4. replace()
print(s.replace('Morph', 'Master'))  # Replaces 'Morph' with 'Master'
# 5. find()
print(s.find('Morph'))  # Returns the index of 'Morph' in the string, or -1 if not found
# 6. rstrip()
print(s.rstrip())  # Removes trailing spaces from the string
# 7. split()
print(s.split())  # Splits the string into a list at spaces
# 8. startswith()
print(s.startswith('Skill'))  # Checks if the string starts with 'Skill'
# 9. isnumeric()
print(s.isnumeric())  # Checks if all characters in the string are numeric (returns False)
# 10. index()
print(s.index('Morph'))  # Returns the index of the first occurrence of 'Morph'
```
