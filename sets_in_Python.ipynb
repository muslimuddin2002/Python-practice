```python
# set {}
my_set = {1, 2, 3, 4}

print(my_set)
```

```python
# Duplicate Value Automatically Remove
numbers = {1, 2, 2, 3, 4, 4}

print(numbers)
```

```python
# Empty Set
my_set = set()

print(type(my_set))
```

```python
# add()
fruits = {"apple", "banana"}

fruits.add("mango")

print(fruits)
```

```python
# Multiple Value Add
# update()
numbers = {1, 2, 3}

numbers.update([4, 5, 6])

print(numbers)
```

```python
# Value Remove
fruits = {"apple", "banana", "mango"}

fruits.remove("banana")

print(fruits)
```

```python
# discard()
fruits = {"apple", "banana"}

fruits.discard("orange")

print(fruits)
```

```python
# Loop Through Set
colors = {"red", "green", "blue"}

for color in colors:
    print(color)
```

```python
# Set Operations 🔥
# Union
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)
```

```python
# Intersection
a = {1, 2, 3}
b = {2, 3, 4}

print(a & b)
```

```python
# Difference
a = {1, 2, 3}
b = {2, 3, 4}

print(a - b)
```

```python
# Real Life Example 🔥
# Unique Student IDs
student_ids = {101, 102, 103, 101, 102}

print(student_ids)
```
