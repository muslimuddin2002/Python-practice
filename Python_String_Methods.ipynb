```python
# String Indexing
# String এর প্রতিটি character এর একটি position থাকে।

text = "Python"

print(text[0])
print(text[1])
print(text[5])
```

```python
# Negative Indexing
text = "Python"

print(text[-1])
print(text[-2])
```

```python
# String Slicing
# String এর নির্দিষ্ট অংশ বের করতে slicing ব্যবহার হয়।
text = "Programming"

print(text[0:6])
print(text[3:8])
```

```python
# Common String Methods
# upper()
# সব letter বড় হাতের করে।
name = "python"

print(name.upper())
```

```python
# lower() সব ছোট হাতের করে।
name = "PYTHON"

print(name.lower())
```

```python
# title() প্রতিটি word এর প্রথম letter capital করে।
text = "i love python"

print(text.title())
```

```python
# strip() শুরুর ও শেষের extra space remove করে।
text = "   Python   "

print(text.strip())
```

```python
# replace() একটি word এর জায়গায় অন্য word বসায়।
text = "I love Java"

print(text.replace("Java", "Python"))
```

```python
# split() String কে list এ ভাগ করে।
text = "Apple Mango Banana"

print(text.split())
```

```python
# find() কোন character বা word কোথায় আছে তা খুঁজে বের করে।
text = "Python Programming"

print(text.find("Program"))
```

```python
# count() কতবার আছে তা গুনে।
text = "apple"

print(text.count("p"))
```

```python
# String Length
text = "Python"

print(len(text))
```
