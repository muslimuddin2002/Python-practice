```python
# Queue Implementation using Linked List 🔥
# Node Class
class Node:

    def __init__(self, value):
        self.value = value
        self.next_node = None

    def get_value(self):
        return self.value

    def get_next_node(self):
        return self.next_node

    def set_next_node(self, next_node):
        self.next_node = next_node


# Queue Class
class Queue:

    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0


    # Enqueue Operation
    def enqueue(self, value):

        item_to_add = Node(value)

        print("Adding", item_to_add.get_value(), "to the queue!")

        if self.is_empty():

            self.head = item_to_add
            self.tail = item_to_add

        else:

            self.tail.set_next_node(item_to_add)
            self.tail = item_to_add

        self.size += 1


    # Dequeue Operation
    def dequeue(self):

        if self.get_size() > 0:

            item_to_remove = self.head

            print(item_to_remove.get_value(), "is served!")

            if self.get_size() == 1:

                self.head = None
                self.tail = None

            else:

                self.head = self.head.get_next_node()

            self.size -= 1

            return item_to_remove.get_value()

        else:
            print("The queue is totally empty!")


    # Peek Operation
    def peek(self):

        if self.size > 0:
            return self.head.get_value()

        else:
            print("No orders waiting!")


    # Queue Size
    def get_size(self):
        return self.size


    # Check Empty
    def is_empty(self):
        return self.size == 0


# ================= TEST =================

pizza_orders = Queue()

pizza_orders.enqueue("Pizza 1")
pizza_orders.enqueue("Pizza 2")
pizza_orders.enqueue("Pizza 3")

print("Front Order:", pizza_orders.peek())

pizza_orders.dequeue()

print("Queue Size:", pizza_orders.get_size())
```
