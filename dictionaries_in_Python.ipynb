```python
#  dictionaries in Python.
student = {
    "name": "Muslim Uddin",
    "department": "CSE",
    "cgpa": 3.85
}

print(student)
```

```python
# Value Access
student = {
    "name": "Muslim",
    "department": "CSE"
}

print(student["name"])
```

```python
# get() Method
student = {
    "name": "Muslim",
    "department": "CSE"
}

print(student.get("department"))
```

```python
# Value Change
student = {
    "name": "Muslim",
    "cgpa": 3.50
}

student["cgpa"] = 3.90

print(student)
```

```python
# New Item Add
student = {
    "name": "Muslim"
}

student["department"] = "CSE"

print(student)
```

```python
# Item Delete
student = {
    "name": "Muslim",
    "department": "CSE"
}

student.pop("department")

print(student)
```

```python
# Dictionary Loop
student = {
    "name": "Muslim",
    "department": "CSE",
    "cgpa": 3.85
}

for key, value in student.items():
    print(key, ":", value)
```
