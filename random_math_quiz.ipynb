```python
# Random Math Quiz

import random

a = random.randint(1, 20)
b = random.randint(1, 20)

print(f"What is {a} + {b} ?")

answer = int(input("Your answer: "))

if answer == a + b:
    print("Correct!")
else:
    print("Wrong!")
    print("Correct answer was:", a + b)
```
