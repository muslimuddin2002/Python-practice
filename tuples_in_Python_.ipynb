```python
#  tuples in Python.
# tuple
my_tuple = ('abc', 123, 'def', 456)

print(my_tuple)
```

```python
# tuple indexing
my_tuple =('abc',1213,'def',4569)
print(my_tuple[0])
print(my_tuple[1])
```

```python
# Negative Indexing
my_tuple = ('abc', 123, 'def', 456)

print(my_tuple[-1])
```

```python
# Tuple Slicing
my_tuple = ('a', 'b', 'c', 'd', 'e')

print(my_tuple[1:4])
```

```python
# Tuple Change করা যায় না
my_tuple = (1, 2, 3)

my_tuple[0] = 100
```

```python
# count()
numbers = (1, 2, 3, 1, 1)

print(numbers.count(1))
```

```python
# index()
numbers = (10, 20, 30)

print(numbers.index(20))
```

```python
# tuple LOop
fruits = ('apple', 'banana', 'mango')

for fruit in fruits:
    print(fruit)
```
