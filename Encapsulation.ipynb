```python
# Encapsulation
class UserInfo:
   def __init__(self, username, email_address):
      self.username = username
      self.email_address = email_address

   def check_username(self, username_to_check):
       if username_to_check == self.username:
           return True
       else:
           return False

user = UserInfo('user123', 'abc@edf.ghi')

print(user.check_username('user123')) # returns True
print(user.check_username('user456')) # returns False
```

```python
# Real Life Encapsulation Example 🔥
# Bank Account System
class BankAccount:

    def __init__(self, owner, balance, pin):

        # Public variable
        self.owner = owner

        # Private variables
        self.__balance = balance
        self.__pin = pin


    # Deposit money
    def deposit(self, amount):

        if amount > 0:
            self.__balance += amount
            print(f"{amount} টাকা জমা হয়েছে")

        else:
            print("Invalid amount")


    # Withdraw money
    def withdraw(self, amount, pin):

        if pin != self.__pin:
            print("Wrong PIN")

        elif amount > self.__balance:
            print("Insufficient balance")

        else:
            self.__balance -= amount
            print(f"{amount} টাকা উত্তোলন হয়েছে")


    # Show balance
    def show_balance(self, pin):

        if pin == self.__pin:
            print("Current Balance:", self.__balance)

        else:
            print("Incorrect PIN")


    # Change PIN
    def change_pin(self, old_pin, new_pin):

        if old_pin == self.__pin:
            self.__pin = new_pin
            print("PIN changed successfully")

        else:
            print("Wrong old PIN")


# Create account
acc1 = BankAccount("Muslim uddin", 5000, 1234)

print("Account Holder:", acc1.owner)

print()

# Deposit money
acc1.deposit(2000)

print()

# Withdraw money
acc1.withdraw(1000, 1234)

print()

# Check balance
acc1.show_balance(1234)

print()

# Change PIN
acc1.change_pin(1234, 5678)

print()

# Check balance with new PIN
acc1.show_balance(5678)
```

```python
# 🏦 Encapsulation Bank System (User Input Version)
class BankAccount:

    def __init__(self, owner, balance, pin):

        self.owner = owner
        self.__balance = balance
        self.__pin = pin


    def deposit(self, amount):

        if amount > 0:
            self.__balance += amount
            print(f"{amount} টাকা জমা হয়েছে")
        else:
            print("Invalid amount")


    def withdraw(self, amount, pin):

        if pin != self.__pin:
            print("Wrong PIN")

        elif amount > self.__balance:
            print("Insufficient balance")

        else:
            self.__balance -= amount
            print(f"{amount} টাকা উত্তোলন হয়েছে")


    def show_balance(self, pin):

        if pin == self.__pin:
            print("Current Balance:", self.__balance)
        else:
            print("Incorrect PIN")


    def change_pin(self, old_pin, new_pin):

        if old_pin == self.__pin:
            self.__pin = new_pin
            print("PIN changed successfully")
        else:
            print("Wrong old PIN")


# ---------------- USER INPUT SECTION ---------------- #

print("=== Bank Account Create ===")

name = input("Enter your name: ")
balance = int(input("Enter initial balance: "))
pin = int(input("Set your PIN: "))

acc = BankAccount(name, balance, pin)

print("\nAccount created successfully!\n")


while True:

    print("\n===== MENU =====")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Show Balance")
    print("4. Change PIN")
    print("5. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        amount = int(input("Enter deposit amount: "))
        acc.deposit(amount)

    elif choice == "2":
        amount = int(input("Enter withdraw amount: "))
        pin = int(input("Enter PIN: "))
        acc.withdraw(amount, pin)

    elif choice == "3":
        pin = int(input("Enter PIN: "))
        acc.show_balance(pin)

    elif choice == "4":
        old_pin = int(input("Enter old PIN: "))
        new_pin = int(input("Enter new PIN: "))
        acc.change_pin(old_pin, new_pin)

    elif choice == "5":
        print("Thank you for using our system 😊")
        break

    else:
        print("Invalid option! Try again")
```

```python
#  🔥 FULL OOP PROJECT CODE
import json
import os


class BankAccount:

    def __init__(self, name, pin, balance=0):

        self.name = name
        self.__pin = pin
        self.__balance = balance
        self.transactions = []


    # ---------------- SECURITY CHECK ----------------
    def __check_pin(self, pin):
        return pin == self.__pin


    # ---------------- DEPOSIT ----------------
    def deposit(self, amount):

        if amount > 0:
            self.__balance += amount
            self.transactions.append(f"Deposited {amount}")
            print(f"{amount} টাকা জমা হয়েছে")
        else:
            print("Invalid amount")


    # ---------------- WITHDRAW ----------------
    def withdraw(self, amount, pin):

        if not self.__check_pin(pin):
            print("Wrong PIN")
            return

        if amount > self.__balance:
            print("Insufficient balance")
            return

        self.__balance -= amount
        self.transactions.append(f"Withdraw {amount}")
        print(f"{amount} টাকা উত্তোলন হয়েছে")


    # ---------------- BALANCE ----------------
    def show_balance(self, pin):

        if self.__check_pin(pin):
            print("Current Balance:", self.__balance)
        else:
            print("Incorrect PIN")


    # ---------------- CHANGE PIN ----------------
    def change_pin(self, old_pin, new_pin):

        if self.__check_pin(old_pin):
            self.__pin = new_pin
            print("PIN changed successfully")
        else:
            print("Wrong old PIN")


    # ---------------- TRANSACTIONS ----------------
    def show_transactions(self, pin):

        if self.__check_pin(pin):
            print("\nTransaction History:")
            for t in self.transactions:
                print("-", t)
        else:
            print("Incorrect PIN")


    # ---------------- SAVE TO FILE ----------------
    def save_data(self):

        data = {
            "name": self.name,
            "pin": self.__pin,
            "balance": self.__balance,
            "transactions": self.transactions
        }

        with open(self.name + ".json", "w") as f:
            json.dump(data, f)


    # ---------------- LOAD FROM FILE ----------------
    @staticmethod
    def load_account(filename):

        if not os.path.exists(filename):
            print("Account not found")
            return None

        with open(filename, "r") as f:
            data = json.load(f)

        acc = BankAccount(data["name"], data["pin"], data["balance"])
        acc.transactions = data["transactions"]

        return acc


# ================= MAIN SYSTEM ================= #

print("=== SMART BANK SYSTEM ===")

choice = input("1. Create Account\n2. Load Account\nChoose: ")

if choice == "1":

    name = input("Enter name: ")
    pin = int(input("Set PIN: "))
    balance = int(input("Initial deposit: "))

    acc = BankAccount(name, pin, balance)

    print("\nAccount Created Successfully!\n")

else:

    filename = input("Enter filename (name.json): ")
    acc = BankAccount.load_account(filename)

    if acc is None:
        exit()


while True:

    print("\n===== MENU =====")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Change PIN")
    print("5. Transactions")
    print("6. Save & Exit")

    option = input("Choose: ")

    if option == "1":
        amount = int(input("Enter amount: "))
        acc.deposit(amount)

    elif option == "2":
        amount = int(input("Enter amount: "))
        pin = int(input("Enter PIN: "))
        acc.withdraw(amount, pin)

    elif option == "3":
        pin = int(input("Enter PIN: "))
        acc.show_balance(pin)

    elif option == "4":
        old = int(input("Old PIN: "))
        new = int(input("New PIN: "))
        acc.change_pin(old, new)

    elif option == "5":
        pin = int(input("Enter PIN: "))
        acc.show_transactions(pin)

    elif option == "6":
        acc.save_data()
        print("Data saved. Goodbye!")
        break

    else:
        print("Invalid option")
```

```python
# 🎓 Encapsulation Example: Student Result System
class StudentResult:

    def __init__(self, name, roll):

        # Public
        self.name = name
        self.roll = roll

        # Private (Encapsulation)
        self.__math = 0
        self.__science = 0
        self.__english = 0


    # Add marks (controlled access)
    def set_marks(self, math, science, english):

        if 0 <= math <= 100 and 0 <= science <= 100 and 0 <= english <= 100:

            self.__math = math
            self.__science = science
            self.__english = english

            print("Marks added successfully")

        else:
            print("Invalid marks! Must be 0-100")


    # Calculate total
    def total(self):

        return self.__math + self.__science + self.__english


    # Calculate percentage
    def percentage(self):

        return self.total() / 3


    # Grade system
    def grade(self):

        avg = self.percentage()

        if avg >= 80:
            return "A+"
        elif avg >= 70:
            return "A"
        elif avg >= 60:
            return "A-"
        elif avg >= 50:
            return "B"
        else:
            return "Fail"


    # Show result
    def show_result(self):

        print("Name:", self.name)
        print("Roll:", self.roll)
        print("Total Marks:", self.total())
        print("Percentage:", self.percentage())
        print("Grade:", self.grade())


# Create student object
s1 = StudentResult("Muslim", 1011)

print()

# Add marks
s1.set_marks(85, 90, 80)

print()

# Show result
s1.show_result()
```
